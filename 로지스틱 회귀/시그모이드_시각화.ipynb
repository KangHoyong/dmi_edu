```python
import numpy as np
import matplotlib.pylab as plt

# [-10, 10] 구간안에 100개의 t값을 시그모이드 함수에 대입한다.
t = np.linspace(-10 , 10, 100) 
# 시그 모이드 공식 
# NumPy의 np.exp() 함수는 밑(base)이 자연상수 e인 지수함수로 변환해줍니다.
sig = 1 / (1 + np.exp(-t)) 

# t 와 시그모이드 결과값을 plotting 
plt.figure(figsize=(9, 3))
plt.plot([-10, 10], [0, 0], "k-")
plt.plot([-10, 10], [0.5, 0.5], "k:") # y축 기준 0.5 실선하나 생성 
plt.plot([-10, 10], [1, 1], "k:") # y 축 기준 1.0 실선하나 생성
plt.plot([0, 0], [-1.1, 1.1], "k-") # center 기준 0.0 선하나 생성 
plt.plot(t, sig, "r-", linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$") # Text를 보여주기위한 용도
plt.xlabel("t") # x 축 name 
plt.legend(loc="upper left", fontsize=20) # Text 크기와 위치 지정 
plt.axis([-10, 10, -0.1, 1.1]) # 그래프 간격 조정

plt.show() # 완성된 그래프 보여주기 
```

```python
# 위와 같이 시그모이드 함수는 S 자 형태를 띄며, 어떤 값(t)을 대입하면 0~1 사이의 값을 반환하도록 해준다.
```

```python
# w값이 변화에 따른 경사도 변화
def sigmoid(t): # 시그모이드 함수 정의
    
    return 1/(1 + np.exp(-t)) 

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5 * x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

print("\n y1 sigmoid \n " , y1)
print("\n y2 sigmoid \n " , y2)
print("\n y3 sigmoid \n " , y3)

plt.plot(x, y1 , 'r', linestyle = '--') # w의 값이 0.5일때
plt.plot(x, y2 , 'g') # w의 값이 1일때
plt.plot(x, y3 , 'b', linestyle = '--') # w의 값이 2일때 
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
# 위의 그래프는 w값이 0.5일경우는 빨강색 W값이 1일때는 초록색선 W의 값이 2일때는 파란색선이 나오도록 하였습니다.
# 자세히 보면 W의 값에 따라 그래프의 경사도가 변하는 것을 볼 수 있습니다.

# 앞서 선형회귀에서 가중치 W는 직선의 기울기를 의미했지만 여기서는 그래프의 경사도를 결정합니다 
# 즉 W의 값이 커지면 경사가 커지고 W의 값이 작아지면 경사도 작아집니다.
```

```python
# b값의 변화에 따른 좌, 우 이동
# 이제 b의 값에 따라서 그래프가 어떻게 변하는지 확인해보겠습니다.
def sigmoid(t): # 시그모이드 함수 정의
    
    return 1/(1 + np.exp(-t)) 

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

print("\n y1 sigmoid \n " , y1)
print("\n y2 sigmoid \n " , y2)
print("\n y3 sigmoid \n " , y3)

plt.plot(x, y1, 'r', linestyle='--') # x + 0.5
plt.plot(x, y2, 'g') # x + 1
plt.plot(x, y3, 'b', linestyle='--') # x + 1.5
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
# 위의 그래프 b의 값에 따라서 그래프가 좌, 우로 이동하는 것을 보여줍니다.

# 이것을가지고 양성 범주에에 속할 확률을 추정하면 이에 대한 예측 y_hat을 쉽게 구할 수 있다.
# 양상(존재여부) 범주 
# 0(음성) if p < 0.5
# 1(양성) if p >= 0.5
# 결론적으로는 t>= 0 이면 sigma(t) >= 0.5 이므로 (반대는 반대로)
# 시그모이드 공식과 함께 생각하면 θ^T·x가 양수일때 1(양성 범주), 음수일때 0(음성 범주) 이라고 예측한다.
```
